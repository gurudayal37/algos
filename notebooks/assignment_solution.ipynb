**Q1: Write a customized function to calculate the factorial of a number. Do not use any
inbuilt functions. Create your own.**

```python
def factorial(x):
    fact = 1;
    if x < 2:
        return fact
    for i in range(1,x+1):
        fact = fact*i
    return fact
```

```python
factorial(5)
```

```python
factorial(10)
```

```python
factorial(1)
```

```python
factorial(0)
```

**Q2: Write a Python program which iterates through a sequence of 50 randomly generated
integers between 100 and 200, and derive two different series/arrays/lists from it, one
which contains even multiples of 3 and the other which contains odd multiples of 5.**

```python
import numpy as np

randomArray = np.random.randint(100, 200, 50)

evenMultipleOfThree = []
oddMultipleOfFive = []

for x in randomArray:
    if( x%3==0 and x%2==0):
        evenMultipleOfThree.append(x)

    if( x%5==0 and x%2==1):
        oddMultipleOfFive.append(x)

print(evenMultipleOfThree)
print(oddMultipleOfFive)
```

**Q3: Write a Python program to generate a sequence of 1000 numbers drawn from a
standard uniform distribution and plot it on a chart. Compute its mean. Use the np.where
clause to create a sequence of only numbers greater than 0.75 from the list. Plot a chart of
those numbers. Check how many such numbers exist. Is it on expected lines? Comment.
Please use np.random.seed(1234) at the beginning of your code.**

```python
import matplotlib.pyplot as plt

np.random.seed(1234) 
uniform_distribution = np.random.rand(1000)
mean = np.mean(uniform_distribution)
print('mean of uniform distribution : ',mean)

new_list = uniform_distribution[np.where(uniform_distribution > 0.75)]

print('new list size : ',len(new_list))

plt.plot(new_list)
plt.title("Values > 0.75 from Uniform Distribution")
plt.xlabel("Index (Filtered)")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
